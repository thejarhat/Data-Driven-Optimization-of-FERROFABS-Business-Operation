```python
X = df_mm.drop(['Burned transformers 2019'],axis=1) 
Y=df_mm["Burned transformers 2019"]
```

```python
Y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
#!pip install imblearn using smote to replicate minority class to fix imbalance
from imblearn.over_sampling import SMOTE 
smote = SMOTE()
X_train_new, y_train_new = smote.fit_resample(X_train, y_train.values.ravel())
import matplotlib.ticker as mtick

# to demonstrate the effect of SMOTE over imbalanced datasets
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize =(15, 5))
ax1.set_title('Before SMOTE')
pd.Series(y_train).value_counts().plot.pie(autopct='%.1f%%',ax=ax1)
ax1.yaxis.set_major_formatter(mtick.PercentFormatter())
ax2.set_title('After SMOTE')  
pd.Series(y_train_new).value_counts().plot.pie(autopct='%.1f%%',ax=ax2)
ax2.yaxis.set_major_formatter(mtick.PercentFormatter())
plt.show()
```

```python
from sklearn.feature_selection import RFE
from sklearn.ensemble import RandomForestClassifier
logreg = RandomForestClassifier()
rfe = RFE(logreg)
rfe = rfe.fit(X_train_new, y_train_new)
```

```python
X=df_mm.drop(['LOCATION', 'Burning rate  [Failures/year]',
       'Criticality according to previous study for ceramics level',
       'Removable connectors', 'Air network', 'Circuit Queue',
       'Type of clients', 'Type of installation','Burned transformers 2019'], axis=1)
Y=df_mm["Burned transformers 2019"]
print(X.shape)
print(Y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=0)
```

```python
from imblearn.over_sampling import SMOTE 
smote = SMOTE()
X_train_new, y_train_new = smote.fit_resample(X_train, y_train.values.ravel())
import matplotlib.ticker as mtick

# to demonstrate the effect of SMOTE over imbalanced datasets
fig, (ax1, ax2) = plt.subplots(ncols = 2, figsize =(15, 5))
ax1.set_title('Before SMOTE')
pd.Series(y_train).value_counts().plot.pie(autopct='%.1f%%',ax=ax1)
ax1.yaxis.set_major_formatter(mtick.PercentFormatter())
ax2.set_title('After SMOTE')  
pd.Series(y_train_new).value_counts().plot.pie(autopct='%.1f%%',ax=ax2)
ax2.yaxis.set_major_formatter(mtick.PercentFormatter())
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression # Logistic Regression
from sklearn.neighbors import KNeighborsClassifier # K-Nearest Neighbbors
from sklearn.naive_bayes import GaussianNB # Gaussain Naive Bayes
from sklearn.tree import DecisionTreeClassifier # Decision Tree
from sklearn.ensemble import RandomForestClassifier # Random Forest
```

```python

lr =  LogisticRegression()
lr.fit(X_train_new, y_train_new)
y_pred=lr.predict(X_test)
```

```python
nb =  GaussianNB()
nb.fit(X_train_new, y_train_new)
y_pred2=nb.predict(X_test)
```

```python
rfm = RandomForestClassifier(n_estimators=70)
rfm.fit(X_train_new, y_train_new)
y_pred3=rfm.predict(X_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier  
classifier= KNeighborsClassifier(n_neighbors=5)  
classifier.fit(X_train_new, y_train_new)  
```

```python
y_pred4=classifier.predict(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model = Sequential()
model.add(Dense(128, input_dim=7, activation='relu'))
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history=model.fit(x=X_train_new,y=y_train_new,epochs=50,validation_data= (X_test, y_test))
```

```python
score = model.evaluate(x=X_test,y=y_test)

print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy Logistic Regression:",metrics.accuracy_score(y_test, y_pred))
print("Accuracy Naive Bayes:",metrics.accuracy_score(y_test, y_pred2))
print("Accuracy Random Forest:",metrics.accuracy_score(y_test, y_pred3))
print("Accuracy KNN:",metrics.accuracy_score(y_test, y_pred4))
print("Accuracy NN:",score[1])
```
